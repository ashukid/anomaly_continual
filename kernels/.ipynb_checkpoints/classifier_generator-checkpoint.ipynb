```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import glob
import shutil
import torch
import torch.nn.functional as F
from torch import nn
import torch.utils.data
import torch.optim as optim
import random
import time

def seed_everything(seed=42):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
seed_everything()
```

```python
class ECGDataset(torch.utils.data.Dataset):
    
    def __init__(self,path):
        self.data=pd.read_csv(path,header=None)
        
    def __getitem__(self,idx):
        x=self.data.loc[idx,:186].values # removed the label
        return x
    
    def __len__(self):
        return len(self.data)
```

```python
previous="mit"
original="inhouse"
```

```python
root_path="../input/ecgdatasets/classification/"+previous
new_path="../input/ecgdatasets/classification/"+original
cuda=torch.cuda.is_available()
device = torch.device("cuda:0" if cuda else "cpu")


class opt:
    batch_size=64
    workers=2
    lr=0.001
    normal_train_path=root_path+"/train.csv"
    new_train_path=new_path+"/train.csv"


df_normal_train=pd.read_csv(opt.new_train_path) #188 columns , last column class = 0
total_data=len(df_normal_train)

generator_ration=0.4
new_data=int(total_data*0.4)
old_data=total_data-new_data

normal_train_dset=ECGDataset(opt.normal_train_path)
trainloader=torch.utils.data.DataLoader(normal_train_dset,
                                       batch_size=new_data,
                                       shuffle=True,
                                        num_workers=opt.workers)

for x in trainloader:
    print(x.shape,x.dtype)
    x=x.numpy()
    x=pd.DataFrame(x)
    x.iloc[:10,:].T.plot(figsize=(20,8))
    plt.show() #plot to check the data pipeline is working well
    break
```

```python
#change this
from models import LSTMEncoder,LSTMClassifier
from models import loss_function
old_network=LSTMEncoder().to(device)
old_network.load_state_dict(torch.load("../models/generator_"+previous+".pt",map_location="cpu"))        
        
old_network.eval()
for i,signal in enumerate(trainloader):
    signal=signal.to(device).float()#batch_size,187
    old_network.zero_grad()
    out1,out2=old_network.Encoder(signal)
    break

    
a,b=new_data,20
r1,r2=out1.min(),out1.max()
noise1 = (r1 - r2) * torch.rand(a, b) + r2

a,b=new_data,20
r1,r2=out2.min(),out2.max()
noise2 = (r1 - r2) * torch.rand(a, b) + r2

out=old_network.Decoder(out1,out2)
out=torch.flip(out,dims=[1])
out=pd.DataFrame(out.detach().numpy())
out.iloc[:10,:].T.plot(figsize=(20,8))
```

```python
# putting labels
solver=LSTMClassifier().to(device)
solver.load_state_dict(torch.load("../models/classifier_"+previous+".pt",map_location="cpu"))        
solver.eval()

labels=[]
for i in range(len(out)):
    x=out.iloc[i,:]
    x=torch.tensor(x).to(device).float()
    x=x.unsqueeze((0))
    label=solver(x)
    label=1 if label>0.5 else 0
    labels.append(label)
    
out["187"]=labels
```

```python
#combining old data with new data
# append a column to out
# combine both dataset and save


##shuffle the data
##label class doesn't matter

df_merged=pd.DataFrame(np.vstack((df_normal_train.values,out.values)))
df_merged = df_merged.sample(frac=1).reset_index(drop=True)
df_merged.to_csv("../input/ecgdatasets/classification/"+previous+"_"+original+"/train.csv",index=False)
print(df_merged.shape)

# mixing validation data
old_valid=pd.read_csv("../input/ecgdatasets/classification/"+previous+"/test.csv")
new_valid=pd.read_csv("../input/ecgdatasets/classification/"+original+"/test.csv")

valid_merge=pd.DataFrame(np.vstack((old_valid.values[:100],new_valid.values[:100])))
valid_merge = valid_merge.sample(frac=1).reset_index(drop=True)
valid_merge.to_csv("../input/ecgdatasets/classification/"+previous+"_"+original+"/test.csv",index=False)
print(valid_merge.shape)
```
